````python
import pandas as pd

df = pd.read_csv('/content/inmigrantes_canada.csv')
df.head()
````

````python
# Ver cantidad de registros
df.shape
````

````python
# Se observa si hay valores nulos o tipos de datos a ajustar, en este caso los datos estan bien
df.info()
````

````python
# Conocer tendencia de migracion de Colombiano a Canada
# Configurar como indice el Pais para poder leer mas facil el conjunto de datos y hacer la modificacion sobre el mismo datafrade inplace=  True
df.set_index('Pais',inplace=True)
df
````

````python
#generar nuevo dataframe con años e inmigrantes provenientes de colombia
#generar varoable con el string de las columnas osea de los años
#uso metodo lista para mapear funcion string, para un rango de 1980 inlcusivo 2014 exclusivo

anos = list(map(str,range(1980,2014)))
anos
````

````python
# metodo .loc ayuda a encontrar el Pais Colombia dentro del filtro de años
colombia = df.loc['Colombia',anos]
colombia
````

````python
df.loc['Colombia','1980']
````

````python
# crear data frame para manejar mejor los datos, vamos a crear un diccionario con el año del objeto colombia que es un panda series , tomo los indices y los paso a una lista, y para inmigrantes la cantidad de inmigrantes de la misma list series , tomo los valores, y convierto en lista tolist()
# Diccionario con años y cantidad de inmigrantes.
col_dict = {'Año':colombia.index.tolist(),   # para año creo una lista
            'Inmigrantes':colombia.values.tolist()}   # cantidad de inmigrates que viajaron durante estos años
col_dict
````

````python
# Creo un Data Frame llamado datos_col, con el diccionario
datos_col = pd.DataFrame(col_dict)
datos_col.tail()
````

## --Matplotlib
https://matplotlib.org/

matplotlib.pyplot Es un módulo de la biblioteca Matplotlib de Python, diseñado para proporcionar una interfaz basada en estados para crear visualizaciones.

Funciona de forma similar a la interfaz de gráficos de MATLAB.

Características y funcionalidades clave de matplotlib.pyplot:

Interfaz basada en estados:
pyplot Mantiene el concepto de "figura actual" y "Ejes actuales", lo que permite a los usuarios realizar cambios incrementales en un gráfico sin hacer referencia explícita a objetos específicos.

Ejemplo:




```
# Esto tiene formato de código
import matplotlib.pyplot as plt

# Create some data
x = [1, 2, 3, 4]
y = [1, 4, 2, 3]

# Plot the data
plt.plot(x, y)

# Add labels and title
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Simple Line Plot")

# Display the plot
plt.show()
```


## Aula 2: Creando figuras con el Matplotlib

````python
# Generar las variables que me permiten tener el liezno (fig) y la grafica (ax)
fig, ax = plt.subplots(figsize=(8,4))  #metodo de plt llamado subplots me permmite graficar en varios cuadros
ax.plot(datos_col['Año'],datos_col['Inmigrantes']) # información a plota eje X y Y
ax.xaxis.set_major_locator(plt.MultipleLocator(5))  # configurar intervalos y rangos, usando localizador multiple
ax.set_title('Inmigración de colombianos hacia Canadá\n1980 - 2013')
ax.set_xlabel('Año')
ax.set_ylabel('Número de Inmigrantes')
plt.show()

````

````python
# Grafico de Caja
fig, axs = plt.subplots(1,2,figsize=(10,3))  # 1 fila con 2 espacios para graficas

# axs[0]. Grafica de la izquierda   - Grafica de Linea con la tendencia
axs[0].plot(datos_col['Año'],datos_col['Inmigrantes'])
axs[0].set_title('Inmigración de Colombianos hacia Canadá\n1980 - 2013')
axs[0].set_xlabel('Año')
axs[0].set_ylabel('Número de Inmigrantes')
axs[0].xaxis.set_major_locator(plt.MultipleLocator(5))
axs[0].grid()

# axs[1]. Grafica de la derecha       - Grafica de Boxplor
axs[1].boxplot(datos_col['Inmigrantes'])  # Medir la cantidad de colombianos
axs[1].set_title('Boxplot de la inmigración de colombianos\nhacia Canadá 1980 - 2013')
axs[1].set_xlabel('Colombia')
axs[1].set_ylabel('Número de Inmigrantes')
axs[1].grid()

plt.show()
````

````python
# Ya manejamos dos graficas ahora vamos a manejar 4 guafricos
fig, axs = plt.subplots(2,2,figsize=(10,6))
fig.subplots_adjust(hspace=0.5,wspace=0.3)  # ajustar espacio entre los graficos hspace = vertical  wspace = horizontal.
fig.suptitle('Inmigración de los 4 mayores países sudamericanos\nhacia Canadá desde 1980 hasta 2013') # Titulo Superior de las graficas, de los 4 paises suramericanos

axs[0,0].plot(df.loc['Colombia',anos])
axs[0,0].set_title('Colombia')

axs[0,1].plot(df.loc['Brasil',anos])
axs[0,1].set_title('Brasil')

axs[1,0].plot(df.loc['Argentina',anos])
axs[1,0].set_title('Argentina')

axs[1,1].plot(df.loc['Perú',anos])
axs[1,1].set_title('Perú')

# para manejar la misma escala para poder evaluar a nivel visual mejor la informacion en las graficas
ymin = 0   # valor minimo para todas las graficas eje y
ymax = 7000  # valor maximo para todas las graficas eje y

# Traer arreglo multimdimensional a uno bidimensional
for ax in axs.ravel():
  ax.set_ylim(ymin,ymax)

# iterar para mejorar la graficacion eje x años separados cada  años
for ax in axs.flat:
  ax.xaxis.set_major_locator(plt.MultipleLocator(5))
  ax.set_xlabel('Año')
  ax.set_ylabel('Número de Inmigrantes')
  ax.grid()

plt.show()
````

````python
# variedad paleta de colores en seaborn es muy amplia
def generar_grafico(palette): #definir  una función que recibe un parametro palette
  fig, ax = plt.subplots(figsize=(8,4))
  ax = sns.barplot(data=top_10,x='Total',y= top_10.index,hue=top_10.index,orient='h',
                   palette=palette, legend=False) # adiciono variable palette
  ax.set_title('Top 10 países con mayor migración hacia Canadá\n1980 - 2013',
              loc='left',fontsize=16)
  ax.set_xlabel('Número de Inmigrantes',fontsize=14)
  ax.set_ylabel('')

  plt.show()
````

Se puede tambien cambiar el tema, podemos tambien colocar las marcaciones, para que el grafico quede mas completo

````python
def generar_grafico(palette): #definir  una función que recibe un parametro palette
  fig, ax = plt.subplots(figsize=(8,4))
  ax.set_frame_on(False)
  ax.xaxis.set_visible(False)
  ax.tick_params(axis='y',which='both',size=0)
  ax = sns.barplot(data=top_10,x='Total',y= top_10.index,hue=top_10.index,orient='h',
                   palette=palette, legend=False) # adiciono variable palette
  ax.set_title('Top 10 países con mayor migración hacia Canadá\n1980 - 2013',
              loc='left',fontsize=16)
  ax.set_xlabel('Número de Inmigrantes',fontsize=14)
  ax.set_ylabel('')
  for i,j in enumerate(top_10['Total']):  # Poner la marcacion de cantidades por paises
    ax.text(j+20,i,str(j),color='black',fontsize=9,ha='left',va='center')
  plt.show()
````

````python
# Eliminar parametros propios de Matplotlib y ajustarlo para Seaborn
def generar_grafico(palette):
  fig, ax = plt.subplots(figsize=(8,4))
  ax.set(xticklabels=[]) # Se eliminan parametros que grafica numero de inmigrantes, se deja solo el numero

  '''
  La linea anterior remplazo estas 3, y veo de nuevo el fondo gris de la grafica
  ax.set_frame_on(False) # Quitar información redundante quitar frame el recuadro al rededor
  ax.xaxis.set_visible(False) # Que elimine del eje x la numeracion las divisiones de abajo
  ax.tick_params(axis='y',which='both',size=0) # Qitar rallitas en Y
  '''
  ax = sns.barplot(data=top_10,x='Total',y= top_10.index,hue=top_10.index,orient='h',
                   palette=palette, legend=False)
  ax.set_title('Top 10 países con mayor migración hacia Canadá\n1980 - 2013',
              loc='left',fontsize=16)
  ax.set_xlabel('Número de Inmigrantes',fontsize=14)
  ax.set_ylabel('')
  sns.despine() # quitar todos los bordes de la gráfica
  for i,j in enumerate(top_10['Total']):
    ax.text(j+20,i,str(j),color='black',fontsize=9,ha='left',va='center')
  plt.show()
````

````python
# permite trabajar de forma iinteractiva con las gráficas
import plotly.express as px
````

````python
# datos_ Colombia, columnas año y cantidad inmigrantes, base para grsficación
datos_col
````

````python
# generar gráfico de linea
fig = px.line(datos_col,x='Año',y='Inmigrantes',
              title='Inmigración de colombianos hacia Canadá entre 1980 y 2013')

fig.update_traces(line_color='green',line_width=3)
# dar un ancho 800 y altura 400, leyenda del eje x dejarlo girado 45 grados
fig.update_layout(width=800,height=400, xaxis={'tickangle':-45},
                   xaxis_title='Año',yaxis_title='Número de Inmigrantes',
                  font_family='Arial',font_size=14, font_color='grey')
fig.show()
````

````python
# El df Sudamerica, sacar Continente, Regiion, Total y dejar solo Pais y los años, y especificar eje 1 = columnas eje = 1
sudamerica.drop(['Continente','Region','Total'],axis=1,inplace=True)
sudamerica
````

````python
# hacemos transposicion donde cada columna sera paises y cada fila los años
sudamerica = sudamerica.T
sudamerica
````

````python
# generar grafica para mostrar diferentes insigth y ver comportamiento de inmigracion a Canada de sudamericanos
fig = px.line(sudamerica,x=sudamerica.index,y=sudamerica.columns,
              title='Inmigración de sudamericanos hacia Canadá entre 1980 y 2013',
              color='Pais', markers=True) # Color por Paises y tener los marcadores para visualizar cada punto
fig.update_layout(width=800,height=400, xaxis={'tickangle':-45},
                  xaxis_title='Año',yaxis_title='Número de Inmigrantes',
                  font_family='Arial',font_size=14, font_color='grey')

# exportar archivo como html para que pueda interactuar a quien le llegue el archivo
fig.write_html('/content/inmigracion_sudamericanos_canada.html')
fig.show()
````

**Plotly Express (px):**
* Ventajas:
  - Simplicidad y rapidez: Crea gráficos complejos con pocas líneas de código, ideal para exploración rápida de datos y visualizaciones básicas.
  - Consistencia: Ofrece una interfaz uniforme para diferentes tipos de gráficos, facilitando el aprendizaje y uso.
  - Integración con Pandas: Trabaja directamente con DataFrames de Pandas, lo que simplifica el manejo de datos.
  - Visualizaciones interactivas: Permite crear gráficos con animaciones y elementos interactivos.

* Desventajas:
  - Menos flexibilidad: Puede no ser suficiente para visualizaciones muy personalizadas o complejas.
  - Menos control directo: No permite acceder a los elementos individuales de la gráfica con tanta granularidad como graph_objs.

**Plotly Graph Objects (go):**
* Ventajas:
  - Máximo control y personalización: Ofrece acceso a todos los elementos de la gráfica, permitiendo modificar cada detalle.
  - Ideal para visualizaciones complejas: Adecuado para gráficos muy específicos o que requieren una personalización profunda.
  - Programación a bajo nivel: Permite crear figuras complejas mediante la composición de objetos gráficos individuales.

* Desventajas:
  - Más código: Requiere más líneas de código para crear gráficos básicos en comparación con plotly.express.
  - Curva de aprendizaje más pronunciada: La manipulación de objetos gráficos puede ser más compleja para principiantes.

  En resumen:
  * Si necesitas crear gráficos rápidamente y con un nivel de personalización estándar, plotly.express es la opción ideal.
  * Si necesitas un control total sobre la apariencia de la gráfica o vas a crear visualizaciones muy complejas, plotly.graph_objs es la mejor opción.


En muchos casos, puedes incluso combinar ambas bibliotecas. Por ejemplo, puedes usar plotly.express para generar una figura base y luego modificarla con objetos gráficos de plotly.graph_objs.

````python
import plotly.graph_objs as go

# Criando uma figura
fig = go.Figure()

# Adicionando a linha do gráfico e definindo a espessura da linha
fig.add_trace(
    go.Scatter(x=[datos_col['Año'].iloc[0]], y=[datos_col['Inmigrantes'].iloc[0]], mode='lines', name='Inmigrantes', line=dict(width=4))
)

# Definir la configuración del layout
fig.update_layout(
    title=dict(
        text='<b>Inmigración de Colombianos hacia Canadá en el periodo de 1980 a 2013</b>',
        x=0.1,
        xanchor='left',
        font=dict(size=18)
    ),
    xaxis=dict(
        range=[int(datos_col['Año'].iloc[0]), int(datos_col['Año'].iloc[-1])],  # Set the x-axis range to cover all years
        autorange=False,
        title='<b>Año</b>',
        tickmode='linear',  # Use linear tick mode
        dtick=5 # Set tick interval to 5 years
    ),
    yaxis=dict(range=[0, 3000], autorange=False, title='<b>Número de Inmigrantes</b>'),
    updatemenus=[dict(
        type='buttons',
        showactive=False,
        buttons=[dict(
            label='Play',
            method='animate',
            args=[None, {'frame': {'duration': 100, 'redraw': True}, 'fromcurrent': True}]
        )]
    )],
    width=1200,
    height=600
)

# Definir la configuración de la animación
frames = [go.Frame(data=[go.Scatter(x=datos_col['Año'].iloc[:i+1], y=datos_col['Inmigrantes'].iloc[:i+1])]) for i in range(len(datos_col))]
fig.frames = frames

# Mostrar la figura
fig.show()
````

````python
fig.write_html('/content/inmigracion_colombia_canada_animado.html', full_html=True)
````
