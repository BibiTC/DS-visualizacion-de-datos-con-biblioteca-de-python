## Etapa 1: DataFrame con el número de ventas de las 4 tiendas en un periodo de Enero a Diciembre de 2022

La gerencia de la empresa necesita visualizar claramente las tendencias de ventas en cada tienda, para que puedan tomar decisiones estratégicas sobre inventarios y acciones de marketing. Para ello, debes crear cuatro subgráficos dispuestos en dos filas y dos columnas, donde cada subgráfico represente una tienda diferente. En este desafío, cada subgráfico debe presentar un gráfico de líneas que muestre la variación del número de ventas a lo largo de los meses del año.
¡Ahora es el momento de mostrar tus habilidades en análisis de datos y visualización! Para crear el DataFrame con el número de ventas de las tiendas y crear la figura, utiliza la siguiente información:

````python
import pandas as pd
import matplotlib.pyplot as plt
````

````python
#Cargar datos una lista(tiendas para usar comoo indice) y un diccionario (Datos por mes de cada tienda)
tiendas = ['A', 'B', 'C', 'D']
ventas_2022 = {'Ene': [100, 80, 150, 50],
    'Feb': [120, 90, 170, 60],
    'Mar': [150, 100, 200, 80],
    'Abr': [180, 110, 230, 90],
    'May': [220, 190, 350, 200],
    'Jun': [230, 150, 280, 120],
    'Jul': [250, 170, 300, 140],
    'Ago': [260, 180, 310, 150],
    'Sep': [240, 160, 290, 130],
    'Oct': [220, 140, 270, 110],
    'Nov': [400, 220, 350, 190],
    'Dec': [300, 350, 400, 250]}
````

````python
# crear un DataFrame con los datos, utilizando la columna "tiendas" como índice
df = pd.DataFrame(ventas_2022, index=tiendas)
````

````python
df
````

crear una figura con un título general que contenga los cuatro subgráficos, cada uno con su respectivo título y etiquetas en los ejes

````python
fig, axs = plt.subplots(2,2,figsize=(10,6))
fig.subplots_adjust(hspace=0.5,wspace=0.3)  # ajustar espacio entre los graficos hspace = vertical  wspace = horizontal.
fig.suptitle('Tendencias de ventas en el periodo de Enero a Diciembre de 2022 en las tiendas A, B, C y D') # Titulo Superior de las graficas, de los 4 paises suramericanos

axs[0,0].plot(df.loc['A'])
axs[0,0].set_title('Ventas en Tienda A')

axs[0,1].plot(df.loc['B'])
axs[0,1].set_title('Ventas en Tienda B')

axs[1,0].plot(df.loc['C'])
axs[1,0].set_title('Ventas en Tienda C')

axs[1,1].plot(df.loc['D'])
axs[1,1].set_title('Ventas en Tienda D')

# para manejar la misma escala para poder evaluar a nivel visual mejor la informacion en las graficas
ymin = 0   # valor minimo para todas las graficas eje y
ymax = 450  # valor maximo para todas las graficas eje y

# Traer arreglo multimdimensional a uno bidimensional
for ax in axs.ravel():
  ax.set_ylim(ymin,ymax)

# iterar para mejorar la graficacion eje x años separados cada  años
for ax in axs.flat:
  #ax.xaxis.set_major_locator(plt.MultipleLocator(5))
  ax.set_xlabel('Mes')
  ax.set_ylabel('Número de Ventas')
  ax.grid()

plt.show()
````

## Etapa 4: Gráfico interactivo con Plotly

Crear un gráfico animado con Plotly que muestre estos datos. El gráfico debe tener las siguientes características:

Dos líneas: una para Brasil y otra para Argentina.
Un botón "Play" para iniciar la animación, mostrando el aumento o disminución del número de inmigrantes a lo largo de los años.
Las configuraciones de animación deben hacer que las dos líneas se muestren y animen al mismo tiempo.

**Consejos:**

* Crea un DataFrame con los datos de Argentina y no olvides dejar la columna de años como tipo int (entero).
* Utiliza el código proporcionado para Brasil como base y adáptalo para incluir los datos de Argentina.
* Para configurar las animaciones, puedes hacer un bucle for para recorrer el DataFrame datos_brasil y, para cada iteración, crear una nueva lista que contenga dos objetos del tipo go.Scatter, uno para cada país. Luego, cada lista puede ser utilizada para crear un objeto go.Frame, que se agrega a la lista de frames. Finalmente, la lista de frames puede ser asignada al objeto fig, que es la figura del gráfico que se animará. Con esto, cuando se inicie la animación, el gráfico mostrará las dos líneas en movimiento, una para Brasil y otra para Argentina.

### Data Frame Brasil y Argentina

````python
anos = list(map(str,range(1980,2014)))
anos
````

````python
brasil = df.loc['Brasil',anos]
brasil
````

````python
argentina = df.loc['Argentina',anos]
argentina
````

````python
# crear diccionario uniendo la lista de años con la cantidad de inmigrantes
bra_dict = {'Año':brasil.index.tolist(),   # para año creo una lista
            'Inmigrantes':brasil.values.tolist()}   # cantidad de inmigrates que viajaron durante estos años
bra_dict
````

````python
# crear diccionario uniendo la lista de años con la cantidad de inmigrantes
arg_dict = {'Año':argentina.index.tolist(),   # para año creo una lista
            'Inmigrantes':argentina.values.tolist()}   # cantidad de inmigrates que viajaron durante estos años
arg_dict
````

````python
# Creo un Data Frame datos_brasil y datos_argentina
datos_brasil = pd.DataFrame(bra_dict)
datos_brasil.tail()
````

````python
# Creo un Data Frame datos_brasil y datos_argentina
datos_argentina= pd.DataFrame(arg_dict)
datos_argentina.tail()
````

### Gráfico animado utilizando Plotly

````python
import plotly.graph_objs as go
fig = go.Figure()
````

````python

fig.add_trace(
    go.Scatter(x=[datos_brasil['Año'].iloc[0]], y=[datos_brasil['Inmigrantes'].iloc[0]], mode='lines', name='Inmigrantes de Brasil', line=dict(width=4))
)
fig.add_trace(
    go.Scatter(x=[datos_argentina['Año'].iloc[0]], y=[datos_argentina['Inmigrantes'].iloc[0]], mode='lines', name='Inmigrantes de Argentina', line=dict(width=4))
)
````

````python
fig.update_layout(
    title=dict(
        text='<b>Inmigración de Brasil y Argentina a Canadá en el período de 1980 a 2013</b>',
        x=0.1,
        font=dict(size=18)
    ),
    xaxis=dict(
        range=[int(datos_brasil['Año'].iloc[0]), int(datos_brasil['Año'].iloc[-1])],  # Set the x-axis range to cover all years
        autorange=False,
        title='<b>Año</b>',
        tickmode='linear',  # Use linear tick mode
        dtick=5 # Set tick interval to 5 years
    ),
    yaxis=dict(range=[0, 3000], autorange=False, title='<b>Número de inmigrantes</b>'),
    updatemenus=[dict(
        type='buttons',
        showactive=False,
        buttons=[dict(
            label='Play',
            method='animate',
            args=[None, {'frame': {'duration': 100, 'redraw': True}, 'fromcurrent': True}]
        )]
    )],
    width=1200,
    height=600
)
````

````python
frames = []
for i in range(len(datos_brasil)):
    frame_data = [
        go.Scatter(x=datos_brasil['Año'].iloc[:i+1], y=datos_brasil['Inmigrantes'].iloc[:i+1]),
        go.Scatter(x=datos_argentina['Año'].iloc[:i+1], y=datos_argentina['Inmigrantes'].iloc[:i+1])
    ]
    frame = go.Frame(data=frame_data)
    frames.append(frame)
fig.frames = frames
fig.show()
````
