````python
import pandas as pd

df = pd.read_csv('/content/inmigrantes_canada.csv')
df.head()
````

````python
# Ver cantidad de registros
df.shape
````

````python
# Se observa si hay valores nulos o tipos de datos a ajustar, en este caso los datos estan bien
df.info()
````

````python
# Conocer tendencia de migracion de Colombiano a Canada
# Configurar como indice el Pais para poder leer mas facil el conjunto de datos y hacer la modificacion sobre el mismo datafrade inplace=  True
df.set_index('Pais',inplace=True)
df
````

````python
#generar nuevo dataframe con años e inmigrantes provenientes de colombia
#generar varoable con el string de las columnas osea de los años
#uso metodo lista para mapear funcion string, para un rango de 1980 inlcusivo 2014 exclusivo

anos = list(map(str,range(1980,2014)))
anos
````

````python
# metodo .loc ayuda a encontrar el Pais Colombia dentro del filtro de años
colombia = df.loc['Colombia',anos]
colombia
````

````python
df.loc['Colombia','1980']
````

````python
# crear data frame para manejar mejor los datos, vamos a crear un diccionario con el año del objeto colombia que es un panda series , tomo los indices y los paso a una lista, y para inmigrantes la cantidad de inmigrantes de la misma list series , tomo los valores, y convierto en lista tolist()
# Diccionario con años y cantidad de inmigrantes.
col_dict = {'Año':colombia.index.tolist(),   # para año creo una lista
            'Inmigrantes':colombia.values.tolist()}   # cantidad de inmigrates que viajaron durante estos años
col_dict
````

````python
# Creo un Data Frame llamado datos_col, con el diccionario
datos_col = pd.DataFrame(col_dict)
datos_col.tail()
````

## --Matplotlib
https://matplotlib.org/

matplotlib.pyplot Es un módulo de la biblioteca Matplotlib de Python, diseñado para proporcionar una interfaz basada en estados para crear visualizaciones.

Funciona de forma similar a la interfaz de gráficos de MATLAB.

Características y funcionalidades clave de matplotlib.pyplot:

Interfaz basada en estados:
pyplot Mantiene el concepto de "figura actual" y "Ejes actuales", lo que permite a los usuarios realizar cambios incrementales en un gráfico sin hacer referencia explícita a objetos específicos.

Ejemplo:




```
# Esto tiene formato de código
import matplotlib.pyplot as plt

# Create some data
x = [1, 2, 3, 4]
y = [1, 4, 2, 3]

# Plot the data
plt.plot(x, y)

# Add labels and title
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Simple Line Plot")

# Display the plot
plt.show()
```





````python
import matplotlib.pyplot as plt
````

````python
# Año : Eje x , Migrantes: eje Y
plt.figure(figsize=(8,4)) # ajustar tamaño de la figura, pulgadas ancho y alto
plt.plot(datos_col['Año'],datos_col['Inmigrantes'])
plt.xticks(['1980','1985','1990','1995','2000','2005','2010']) # cada rallita eje X
plt.show()
````

````python
# Necesito generar una figura y plot. Hasta el punto anterior nada se esta almmacenando en una variable
plt.figure(figsize=(8,4)) # Donde voy a graficar
plt.plot(datos_col['Año'],datos_col['Inmigrantes'])  # grafica
plt.xticks(['1980','1985','1990','1995','2000','2005','2010'])
plt.title('Inmigración de colombianos hacia Canadá') # Titulo
plt.xlabel('Año')  # Etiqueta Eje X
plt.ylabel('Número de Inmigrantes') # Etiqueta Eje Y
plt.show() # debe ser el ultimo comando
````


## Aula 2: Creando figuras con el Matplotlib
---

````python
# Generar las variables que me permiten tener el liezno (fig) y la grafica (ax)
fig, ax = plt.subplots(figsize=(8,4))  #metodo de plt llamado subplots me permmite graficar en varios cuadros
ax.plot(datos_col['Año'],datos_col['Inmigrantes']) # información a plota eje X y Y
ax.xaxis.set_major_locator(plt.MultipleLocator(5))  # configurar intervalos y rangos, usando localizador multiple
ax.set_title('Inmigración de colombianos hacia Canadá\n1980 - 2013')
ax.set_xlabel('Año')
ax.set_ylabel('Número de Inmigrantes')
plt.show()

````

````python
# Grafico de Caja
fig, axs = plt.subplots(1,2,figsize=(10,3))  # 1 fila con 2 espacios para graficas

# axs[0]. Grafica de la izquierda   - Grafica de Linea con la tendencia
axs[0].plot(datos_col['Año'],datos_col['Inmigrantes'])
axs[0].set_title('Inmigración de Colombianos hacia Canadá\n1980 - 2013')
axs[0].set_xlabel('Año')
axs[0].set_ylabel('Número de Inmigrantes')
axs[0].xaxis.set_major_locator(plt.MultipleLocator(5))
axs[0].grid()

# axs[1]. Grafica de la derecha       - Grafica de Boxplor
axs[1].boxplot(datos_col['Inmigrantes'])  # Medir la cantidad de colombianos
axs[1].set_title('Boxplot de la inmigración de colombianos\nhacia Canadá 1980 - 2013')
axs[1].set_xlabel('Colombia')
axs[1].set_ylabel('Número de Inmigrantes')
axs[1].grid()

plt.show()
````

````python
# Ya manejamos dos graficas ahora vamos a manejar 4 guafricos
fig, axs = plt.subplots(2,2,figsize=(10,6))
fig.subplots_adjust(hspace=0.5,wspace=0.3)  # ajustar espacio entre los graficos hspace = vertical  wspace = horizontal.
fig.suptitle('Inmigración de los 4 mayores países sudamericanos\nhacia Canadá desde 1980 hasta 2013') # Titulo Superior de las graficas, de los 4 paises suramericanos

axs[0,0].plot(df.loc['Colombia',anos])
axs[0,0].set_title('Colombia')

axs[0,1].plot(df.loc['Brasil',anos])
axs[0,1].set_title('Brasil')

axs[1,0].plot(df.loc['Argentina',anos])
axs[1,0].set_title('Argentina')

axs[1,1].plot(df.loc['Perú',anos])
axs[1,1].set_title('Perú')

# para manejar la misma escala para poder evaluar a nivel visual mejor la informacion en las graficas
ymin = 0   # valor minimo para todas las graficas eje y
ymax = 7000  # valor maximo para todas las graficas eje y

# Traer arreglo multimdimensional a uno bidimensional
for ax in axs.ravel():
  ax.set_ylim(ymin,ymax)

# iterar para mejorar la graficacion eje x años separados cada  años
for ax in axs.flat:
  ax.xaxis.set_major_locator(plt.MultipleLocator(5))
  ax.set_xlabel('Año')
  ax.set_ylabel('Número de Inmigrantes')
  ax.grid()

plt.show()
````
