```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = df = pd.read_csv('/content/inmigrantes_canada.csv')
df.head()
```

```python
df.info()
```

```python
df.set_index('Pais',inplace=True)
df
```

```python
anos = list(map(str,range(1980,2014)))
anos
```

```python
# metodo .loc ayuda a encontrar el Pais Colombia dentro del filtro de años
df_brasil_argentina = df.loc[['Brasil','Argentina'],anos]
df_brasil_argentina
```

```python
type(df_brasil_argentina)
```

```python
df_brasil_argentina = df_brasil_argentina.transpose()
# si no uso transpose los años se ven como columnas

df_brasil_argentina.head()

```

# Grafica

-----

```python
# Necesito generar una figura y plot. Hasta el punto anterior nada se esta almmacenando en una variable
plt.figure(figsize=(8,4))
plt.plot(df_brasil_argentina['Brasil'], label='Brasil')
plt.plot(df_brasil_argentina['Argentina'], label='Argentina')
plt.xticks(['1980','1985','1990','1995','2000','2005','2010'])
plt.title('Inmigración de Brasil y Argentina hacia Canadá') # Titulo
plt.xlabel('Año')  # Etiqueta Eje X
plt.ylabel('Número de Inmigrantes') # Etiqueta Eje Y
plt.legend() # Add a legend to identify the lines
plt.show() # debe ser el ultimo comando
```

¿Hay alguna tendencia o patrón común en los datos de ambos países?

La tendencia de los dos Paises Brasil y Argentina fue en aumento y se acelero despues del año 2000. En argentina se freno ese crecimiento en el año 2004 y en Brasil hasta el año 2010, a partir de esos aso comenzo a descender la inmigración hacia Canada

¿Cuáles son los períodos con mayor número de inmigrantes en ambos países?

El periodo con mayor numero de inmigrantes en ambos paises fue de 1990 a 1992,
Luego para Argentina su mayor numero de inmigrantes a Canada fue en 2002, y para Brasil en 2010
