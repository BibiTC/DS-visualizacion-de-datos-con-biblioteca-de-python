## Etapa 1: DataFrame con el número de ventas de las 4 tiendas en un periodo de Enero a Diciembre de 2022

Trabajas como Analista de Datos en una empresa minorista y te asignaron la tarea de crear una figura con subgráficos que muestre la variación en el número de ventas en cuatro tiendas diferentes a lo largo de un año. La gerencia de la empresa necesita visualizar claramente las tendencias de ventas en cada tienda, para que puedan tomar decisiones estratégicas sobre inventarios y acciones de marketing. Para ello, debes crear cuatro subgráficos dispuestos en dos filas y dos columnas, donde cada subgráfico represente una tienda diferente. En este desafío, cada subgráfico debe presentar un gráfico de líneas que muestre la variación del número de ventas a lo largo de los meses del año.
¡Ahora es el momento de mostrar tus habilidades en análisis de datos y visualización! Para crear el DataFrame con el número de ventas de las tiendas y crear la figura, utiliza la siguiente información:

````python
import pandas as pd
import matplotlib.pyplot as plt
````

````python
#Cargar datos una lista(tiendas para usar comoo indice) y un diccionario (Datos por mes de cada tienda)
tiendas = ['A', 'B', 'C', 'D']
ventas_2022 = {'Ene': [100, 80, 150, 50],
    'Feb': [120, 90, 170, 60],
    'Mar': [150, 100, 200, 80],
    'Abr': [180, 110, 230, 90],
    'May': [220, 190, 350, 200],
    'Jun': [230, 150, 280, 120],
    'Jul': [250, 170, 300, 140],
    'Ago': [260, 180, 310, 150],
    'Sep': [240, 160, 290, 130],
    'Oct': [220, 140, 270, 110],
    'Nov': [400, 220, 350, 190],
    'Dec': [300, 350, 400, 250]}
````

````python
type(tiendas)
````

````python
type(ventas_2022)
````

````python
# crear un DataFrame con los datos, utilizando la columna "tiendas" como índice
df = pd.DataFrame(ventas_2022, index=tiendas)
````

````python
df
````

crear una figura con un título general que contenga los cuatro subgráficos, cada uno con su respectivo título y etiquetas en los ejes

````python
fig, axs = plt.subplots(2,2,figsize=(10,6))
fig.subplots_adjust(hspace=0.5,wspace=0.3)  # ajustar espacio entre los graficos hspace = vertical  wspace = horizontal.
fig.suptitle('Tendencias de ventas en el periodo de Enero a Diciembre de 2022 en las tiendas A, B, C y D') # Titulo Superior de las graficas, de los 4 paises suramericanos

axs[0,0].plot(df.loc['A'])
axs[0,0].set_title('Ventas en Tienda A')

axs[0,1].plot(df.loc['B'])
axs[0,1].set_title('Ventas en Tienda B')

axs[1,0].plot(df.loc['C'])
axs[1,0].set_title('Ventas en Tienda C')

axs[1,1].plot(df.loc['D'])
axs[1,1].set_title('Ventas en Tienda D')

# para manejar la misma escala para poder evaluar a nivel visual mejor la informacion en las graficas
ymin = 0   # valor minimo para todas las graficas eje y
ymax = 450  # valor maximo para todas las graficas eje y

# Traer arreglo multimdimensional a uno bidimensional
for ax in axs.ravel():
  ax.set_ylim(ymin,ymax)

# iterar para mejorar la graficacion eje x años separados cada  años
for ax in axs.flat:
  #ax.xaxis.set_major_locator(plt.MultipleLocator(5))
  ax.set_xlabel('Mes')
  ax.set_ylabel('Número de Ventas')
  ax.grid()

plt.show()
````

## Etapa 2: Personalizando los subplots con los datos de ventas de tiendas diferentes

En la Etapa 1, tuviste el desafío de crear una figura con subgráficos que presentan la variación en el número de ventas en cuatro tiendas diferentes a lo largo de un año. ¡Ahora es el momento de llevar esa figura a un nuevo nivel! ¡Es hora de personalizarla! En esta segunda parte del desafío, debes explorar las opciones de personalización de los subgráficos para hacer que la figura sea más clara y atractiva para la gerencia de la empresa.

Algunas ideas de personalización que puedes explorar son:

1. Cambiar la posición de los títulos de los subgráficos a la izquierda



```
loc='left'
```



2. Aumentar el tamaño de la fuente del título general de la figura para resaltarlo



```
fontsize=20
```



3. Aumentar el tamaño de los títulos y etiquetas de los ejes de los subgráficos.



```
fontsize=18
fontsize=14
```



4. Hacer que las líneas sean más gruesas.



```
lw=2.5
```



5. Cambiar el color de las líneas de cada tienda para diferenciarlas aún más.



```
color='g'
color='r'
color='b'
color='p'
```



````python
fig,axs = plt.subplots(2,2,figsize=(10,6))
fig.subplots_adjust(hspace=0.5,wspace=0.3)  # ajustar espacio entre los graficos hspace = vertical  wspace = horizontal.
fig.suptitle('Tendencias de ventas en el periodo de Enero a Diciembre de 2022 \nen las tiendas A, B, C y D',fontsize=18) # Titulo Superior de las graficas, de los 4 paises suramericanos

axs[0,0].plot(df.loc['A'],lw=2.5,color='g')
axs[0,0].set_title('Ventas en Tienda A', fontsize=16, loc='left')

axs[0,1].plot(df.loc['B'],lw=2.5,color='r')
axs[0,1].set_title('Ventas en Tienda B', fontsize=16, loc='left')

axs[1,0].plot(df.loc['C'],lw=2.5,color='b')
axs[1,0].set_title('Ventas en Tienda C', fontsize=16, loc='left')

axs[1,1].plot(df.loc['D'],lw=2.5,color='y')
axs[1,1].set_title('Ventas en Tienda D', fontsize=16, loc='left')

# para manejar la misma escala para poder evaluar a nivel visual mejor la informacion en las graficas
ymin = 0   # valor minimo para todas las graficas eje y
ymax = 450  # valor maximo para todas las graficas eje y

# Traer arreglo multimdimensional a uno bidimensional
for ax in axs.ravel():
  ax.set_ylim(ymin,ymax)

# iterar para mejorar la graficacion eje x años separados cada  años
for ax in axs.flat:
  #ax.xaxis.set_major_locator(plt.MultipleLocator(5))
  #ax.set_xlabel('Mes',fontsize=12)
  ax.set_ylabel('Número de Ventas',fontsize=10)
  ax.spines['top'].set_visible(False)  # quitar lineas del marco superior y derecha
  ax.spines['right'].set_visible(False) # quitar lineas del marco superior y derecha
  #ax.grid()

fig.tight_layout(rect=[0, 0.03, 1, 0.95]) # Ajusto el espacio para que el titulo general no se sobreponga con el titulo de los subgraficos

plt.show()
````

Puedes reducir el tamaño del código utilizando el comando for i, ax in enumerate(axs.flat):, que permite un bucle que itera sobre todos los subgráficos de la figura. Dentro de este bucle, puedes pasar las funciones plot, set_title, set_xlabel, set_ylabel, y así sucesivamente...

¡Ahora es el momento de poner manos a la obra! Experimenta diferentes personalizaciones y haz que la figura sea aún más impresionante.
